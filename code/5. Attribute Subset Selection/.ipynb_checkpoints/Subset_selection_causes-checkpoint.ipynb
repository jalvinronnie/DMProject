```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import math
```

"cal_corr" is user defined function to calculate correlation coefficient.

```python
def cal_corr (p,q,data):
    x=[]
    y=[]
    for i in data[p]:
     x.append(i)
    for i in data[q]:
     y.append(i)
    x = x - np.mean(x)
    y = y - np.mean(y)
    return np.dot(x, y) / math.sqrt(np.dot(x, x) * np.dot(y, y))
```

Reading data from "encoded_causes.xlsx" and storing in "data".

```python
data = pd.read_excel('~/DMFinalTerm_TeamID_05/data/encoded_causes.xlsx',index_col=0)
data
```

"d1" contains the subset of attributes for attribute subset selection.

```python
d1 = ['C1','C2','C3','Male','Female','Total']
d1
```

```python
# ans is used to store the candidte attributes for removal. 
ans=[]
```

Evaluation is done on the correlation coefficient for all objects in "d1" taken pair-wise.
Threshold correlation coefficient is taken as 0.98.
Stopping criterion is "No pair of attributes in d1 exists with correlation coefficient greater than Threshold correlation coefficient."

```python
flag1=0
while flag1 == 0 :
    flag2=0
    for i in d1:
        for j in d1:
            cor=cal_corr(i,j,data)
            if((cor  > 0.98) & (i != j)):
                ans.append([i,j,cor])
                flag2=1
                d1.remove(j)   
    if(flag2 == 0):
        flag1=1
```

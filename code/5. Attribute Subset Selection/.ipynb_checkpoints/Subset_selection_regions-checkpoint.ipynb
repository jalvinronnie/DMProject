```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_excel('~/DMFinalTerm_TeamID_05/data/encoded_regions.xlsx',index_col=0)
data
```

```python
d1 = ['R1','R2','R3','Male','Female','Total']
d1
```

```python
ans = []
```

```python
for i in d1:
    for j in d1:
        cor=np.corrcoef(data[i],data[j])
        if((cor[0][1]  > 0.98) & (i != j)):
         ans.append([i,j,cor[0][1]])
        
```

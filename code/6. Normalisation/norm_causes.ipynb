```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Reading data from "subset_selection_cause.xlsx" and storing in "data".

```python
data = pd.read_excel('~/DMFinalTerm_TeamID_05/data/subset_selection_cause.xlsx')
data
```

Inserting attribute "Male%" and "Female%" with initial attribute values "0".

```python
data.insert(7,'Male%',0)
data.insert(8,'Female%',0)
```

Calculating percentage of Male and inserting them into the attribute "Male%".

```python
for i in data['Index']:
    k=data.loc[data['Index']==i,'Male'].sum()
    if k!=0:
        data.loc[data['Index']==i,'Male%']=data['Male']*100/k
    else:
        data.loc[data['Index']==i,'Male%']=0
```

Calculating percentage of Female and inserting them into the attribute "Female%".

```python
for i in data['Index']:
    k=data.loc[data['Index']==i,'Female'].sum()
    if k!=0:
        data.loc[data['Index']==i,'Female%']=data['Female']*100/k
    else:
        data.loc[data['Index']==i,'Female%']=0
        
```

Renaming attributes "Male%" and "Female%" as "Male" and "Female" respectively.

```python
d=data.drop(['Male','Female'],axis=1)
d.columns=['Index','STATE/UT','Year','C1','C2','C3','Male','Female']
d.set_index('Index',inplace=True)
```

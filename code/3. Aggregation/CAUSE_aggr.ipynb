```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from numpy import*
import math
```

Reading data from "data_cleaning.xlsx".

```python
data=pd.read_excel('~/DMFinalTerm_TeamID_05/data/data_cleaning.xlsx',index_col=0)
```

"data1" contains tuples from "data" which have the following attribute objects under "CAUSE":
1.Illness (Aids/STD)
2.Cancer','Paralysis
3.Insanity/Mental Illness
4.Other Prolonged Illness
5.Drug Abuse/Addiction.

```python
data1=data[data['CAUSE'].isin(['Illness (Aids/STD)','Cancer','Paralysis','Insanity/Mental Illness','Other Prolonged Illness','Drug Abuse/Addiction'])]
```

"a" has the aggregation of attribute ojects:
1.Illness (Aids/STD)
2.Cancer','Paralysis
3.Insanity/Mental Illness
4.Other Prolonged Illness
5.Drug Abuse/Addiction
and is further grouped by 'STATE/UT'and'Year'.

```python

a=data1.groupby(['STATE/UT','Year']).sum().reset_index()
a.insert(1,'CAUSE','Physical/Mental Illness')
```

"data2" contains tuples from "data" which have the following attribute objects under "CAUSE":
1.Bankruptcy or Sudden change in Economic Status
2.Suspected/Illicit Relation
3.Cancellation/Non-Settlement of Marriage
4.Death of Dear Person
5.Dowry Dispute
6.Divorce
7.Fall in Social Reputation
8.Family Problems
9.Love Affairs
10.Poverty
11.Professional/Career Problem
12.Property Dispute','Unemployment

```python
data2=data[data['CAUSE'].isin(['Bankruptcy or Sudden change in Economic Status','Suspected/Illicit Relation','Cancellation/Non-Settlement of Marriage','Death of Dear Person','Dowry Dispute','Divorce','Fall in Social Reputation','Family Problems','Love Affairs','Poverty','Professional/Career Problem','Property Dispute','Unemployment'])]
```

"b" has the aggregation of attribute ojects:
1.Bankruptcy or Sudden change in Economic Status
2.Suspected/Illicit Relation
3.Cancellation/Non-Settlement of Marriage
4.Death of Dear Person
5.Dowry Dispute
6.Divorce
7.Fall in Social Reputation
8.Family Problems
9.Love Affairs
10.Poverty
11.Professional/Career Problem
12.Property Dispute','Unemployment
and is further grouped by 'STATE/UT'and'Year'.

```python
b=data2.groupby(['STATE/UT','Year']).sum().reset_index()
b.insert(1,'CAUSE','Economic/Social Problem')
```

"data3" contains tuples from "data" which have the following attribute objects under "CAUSE":
1.Not having Children(Barrenness/Impotency)
2.Failure in Examination
3.Ideological Causes/Hero Worshipping
4.Illegitimate Pregnancy
5.Physical Abuse (Rape,Incest Etc.)
6.Causes Not known
7.Other Causes (Please Specity)

```python
data3=data[data['CAUSE'].isin(['Not having Children(Barrenness/Impotency)','Failure in Examination','Ideological Causes/Hero Worshipping','Illegitimate Pregnancy','Physical Abuse (Rape,Incest Etc.)','Causes Not known','Other Causes (Please Specity)'])]
```

"c" has the aggregation of attribute ojects:
1.Not having Children(Barrenness/Impotency)
2.Failure in Examination
3.Ideological Causes/Hero Worshipping
4.Illegitimate Pregnancy
5.Physical Abuse (Rape,Incest Etc.)
6.Causes Not known
7.Other Causes (Please Specity)
and is further grouped by 'STATE/UT'and'Year'.

```python
c=data3.groupby(['STATE/UT','Year']).sum().reset_index()
c.insert(1,'CAUSE','Other Causes')
```

"d" is the concatenation of "a","b","c".

```python
d=pd.concat([a,b,c], ignore_index=False)
d.sort_values(by=['STATE/UT','Year'],inplace=True)
```

```python
plt.figure(figsize=(12,12))
```

```python
e=d[d['CAUSE'].isin(['Physical/Mental Illness'])]
sns.lineplot(x=e['Year'],y=e['Male'])
```

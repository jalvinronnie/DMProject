```python
import pandas as pd
import numpy as np
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

Reading data from "merged.xlsx".

```python
data=pd.read_excel('~/DMFinalTerm_TeamID_05/data/merged.xlsx')
```

Data Cleaning:Removing the attribute object "Total" from attribute "CAUSE".

```python
data=data[~data['CAUSE'].isin(['Total','Total Illness'])]
```

Removing redundant attribute objects "TOTAL (ALL INDIA)","TOTAL (STATES)","TOTAL (UTs)".

```python
data=data[~data['STATE/UT'].isin(['TOTAL (ALL INDIA)','TOTAL (STATES)','TOTAL (UTs)'])]
```

```python
d1=data.drop(['Year','CAUSE'],axis=1)
d3=d1.groupby(['STATE/UT']).sum()
```

```python

plt.figure(figsize=(7,7))
font = { 'size'   : 7 }

plt.rc('font', **font)
d3['Total'].plot(kind="barh")
plt.xlabel('Total no. of suicide cases')

plt.show()
```

Exporting "data" as "data_cleaning.xlsx" after data cleaning.

```python
data.to_excel("~/DMFinalTerm_TeamID_05/data/data_cleaning.xlsx")
```

```python
def entropy(target_col):  
    
    elements,counts = np.unique(target_col,return_counts = True)  
    entropy = np.sum([(-counts[i]/np.sum(counts))*np.log2(counts[i]/np.sum(counts)) for i in range(len(elements))])  
    return entropy  
```

```python
def InfoGain(data,split_attribute_name,target_name="Male"):  
         
    #Calculates the entropy of the total dataset  
    total_entropy = entropy(data[target_name])  
      
    #Calculates the entropy of the dataset  
      
    #Calculate the values and the corresponding counts for the split attribute   
    vals,counts= np.unique(data[split_attribute_name],return_counts=True)  
      
    #Calculates the weighted entropy  
    Weighted_Entropy = np.sum([(counts[i]/np.sum(counts))*entropy(data.where(data[split_attribute_name]==vals[i]).dropna()[target_name]) for i in range(len(vals))])  
      
    #Calculates the information gain  
    Information_Gain = total_entropy - Weighted_Entropy  
    return Information_Gain  
```

```python
def ID3(data,originaldata,features,target_attribute_name="Male",parent_node_class = None):  
  
    #Defining the stopping criteria   
      
    
    if len(np.unique(data[target_attribute_name])) <= 1:  
        return np.unique(data[target_attribute_name])[0]  
      
    #If the dataset is empty, return the mode target feature value in the original dataset  
    elif len(data)==0:  
        return np.unique(originaldata[target_attribute_name])[np.argmax(np.unique(originaldata[target_attribute_name],return_counts=True)[1])]  
      
    #If the feature space is empty, return the mode target feature value of the direct parent node 
      
    elif len(features) ==0:  
        return parent_node_class  
      
    #If none of the above holds true, tree is grown 
      
    else:  
        #Setting the default value for this node 
        parent_node_class = np.unique(data[target_attribute_name])[np.argmax(np.unique(data[target_attribute_name],return_counts=True)[1])]  
          
        #Selecting the feature which best splits the dataset  
        item_values = [InfoGain(data,feature,target_attribute_name) for feature in features]
        #Returns the information gain values for the features in the dataset
        
        best_feature_index = np.argmax(item_values)  
        best_feature = features[best_feature_index]  
          
        #Creating the tree structure. 
        tree = {best_feature:{}}  
          
          
        #Removing the feature with the best inforamtion gain from the feature space  
        features = [i for i in features if i != best_feature]  
          
        #Growing a branch under the root node for each possible value of the root node feature  
          
        for value in np.unique(data[best_feature]):  
            value = value  
            #Splitting the dataset along the value of the feature with the largest information gain and therwith create sub_datasets  
            sub_data = data.where(data[best_feature] == value).dropna()  
              
            #Calling the ID3 algorithm for each of those sub_datasets with the new parameters   
            subtree = ID3(sub_data,dataset,features,target_attribute_name,parent_node_class)  
              
            #Adding the sub tree, grown from the sub_dataset to the tree under the root node  
            tree[best_feature][value] = subtree  
              
        return(tree)      
                  
```

```python
def predict(query,tree,default = 1):  
    
    for key in list(query.keys()):  
        if key in list(tree.keys()):  
            
            try:  
                result = tree[key][query[key]]   
            except:  
                return default  
    
            result = tree[key][query[key]]  
            
            if isinstance(result,dict):  
                return predict(query,result)  
            else:  
                return result  
```

```python
def train_test_split(dataset):  
    training_data = dataset.iloc[:80].reset_index(drop=True)#We drop the index respectively relabel the index  
    #starting form 0, because we do not want to run into errors regarding the row labels / indexes  
    testing_data = dataset.iloc[80:].reset_index(drop=True)  
    return training_data,testing_data  
```

```python
training_data = train_test_split(dataset)[0]  
testing_data = train_test_split(dataset)[1] 
```

```python
def test(data,tree):  
    #Creating new query instances by simply removing the target feature column from the original dataset and   
    #convert it to a dictionary  
    queries = data.iloc[:,:-1].to_dict(orient = "records")  
      
    #Create a empty DataFrame in whose columns the prediction of the tree are stored  
    predicted = pd.DataFrame(columns=["predicted"])   
      
    #Calculate the prediction accuracy  
    for i in range(len(data)):  
        predicted.loc[i,"predicted"] = predict(queries[i],tree,1.0)   
    print('The prediction accuracy is: ',(np.sum(predicted["predicted"] == data["Male"])/len(data))*100,'%') 
```

```python
tree = ID3(training_data,training_data,training_data.columns[:-1])  
pprint(tree)  
test(testing_data,tree) 
```
